```python
cols = [
    'id',
    'title',
    'adult',
    'overview',
    'genres',
    'production_companies',
    'production_countries',
    'spoken_languages',
    'keywords'
]

movies_work = movies[cols]
movies_work.head()
```

```python
movies_work.info()
```

```python
movies_work.isnull().sum()
```

```python
movies_work = movies_work.dropna()
```

```python
movies_work.shape
```

```python
def convert(string):
    return [ i.replace(" ", "") for i in str(string).lower().split(',')]
```

```python
convert(movies.iloc[0].production_companies)
```

```python
cols_to_list = [
    'genres',
    'production_companies',
    'production_countries',
    'spoken_languages',
    'keywords'
]

for col in cols_to_list:
    movies_work.loc[:, col] = movies_work[col].apply(convert)
```

```python
movies_work['adult'] = movies_work['adult'].apply(lambda x : ['adult-1'] if x else ['adult-0'])
```

```python
movies_work['overview']=movies_work['overview'].apply(lambda x:str(x).lower().split())
```

```python
movies_work.head()
```

```python
movies_work['tags'] = movies_work['overview']+movies_work['genres']+movies_work['production_companies']+movies_work['production_countries']+movies_work['spoken_languages']+movies_work['keywords']+movies_work['adult']
```

```python
movies_work.head()
```

```python
cols_to_drop = [
    'genres',
    'adult',
    'overview',
    'production_companies',
    'production_countries',
    'spoken_languages',
    'keywords'
]
new_movie = movies_work.drop(columns=cols_to_drop)
```

```python
new_movie.head()
```

```python
new_movie['tags'] = new_movie['tags'].apply(lambda x: " ".join(x))
new_movie.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
vector = cv.fit_transform(new_movie['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new_movie[new_movie['title'] == 'Avatar'].index[0]
```

```python
def recommend(movie):
    index = new_movie[new_movie['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:11]:
        print(new_movie.iloc[i[0]].title)
```

```python
recommend('Batman')
```

```python
recommend('Batman Begins')
```
